```python
from sklearn.model_selection import train_test_split

x = df.drop(columns=["diagnosis"], axis=1)
y = df["diagnosis"]
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(x)
X = scaler.transform(x)
X = pd.DataFrame(X)

x_train, x_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 42)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score

depth = range(1, 35)
test_accuracy = list()
train_accuracy = list()
score = 0

for i in depth:
    tree = DecisionTreeClassifier(max_depth = i, criterion = 'entropy')
    tree.fit(x_train, y_train)
    yhat_train = tree.predict(x_train)
    train_accuracy.append(accuracy_score(y_train, yhat_train))
    
    yhat_test = tree.predict(x_test)
    accuracy = accuracy_score(y_test,yhat_test)
    test_accuracy.append(accuracy)
    
    if score < accuracy:
        score = accuracy
        best_depth = i
        
plt.figure(figsize = (11, 11))
plt.plot(depth, train_accuracy, '-o', label='Train Accuracy')
plt.plot(depth, test_accuracy, '-o', label='Test Accuracy')
plt.legend()
plt.show()
```

```python
print("Based on the plot above, choose max_depth = ", best_depth, "and accuracy from Decision Tree Classifier: ", score*100, "%")
print("The model is overfitting since the train accuracy increased to 1.00 and the test accuracy slightly decreased.")
```
